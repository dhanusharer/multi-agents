## 3. Multilingual Demos

Let's test queries across 5 supported languages: Hindi, Kannada, Telugu, Marathi, and English.

```python
# Demo 1: Crop Advisory in Hindi
query_hi = "मेरी टमाटर की पत्तियां पीली पड़ रही हैं। क्या करूं?"
print(f"User Query (hi): {query_hi}\n")
response_hi = await orchestrator.run(query_hi, session_id="demo_hi")
print(response_hi)
```

```python
# Demo 2: Scheme Eligibility check in Kannada
query_kn = "ನನ್ನ 1.5 ಎಕರ್ ಭೂಮಿ ಇದೆ. ನಾನು PM-KISAN ಗೆ ಅರ್ಹನಾ?"
print(f"User Query (kn): {query_kn}\n")
response_kn = await orchestrator.run(query_kn, session_id="demo_kn")
print(response_kn)
```

```python
# Demo 3: Market Prices & Recommendation in Telugu
query_te = "గుంటూరులో వరి ధర ఎంత? అమ్మాలా?"
print(f"User Query (te): {query_te}\n")
response_te = await orchestrator.run(query_te, session_id="demo_te")
print(response_te)
```

```python
# Demo 4: Weather Planning in Marathi
query_mr = "नाशिक मध्ये टोमॅटो पिकासाठी हवामान कसे आहे?"
print(f"User Query (mr): {query_mr}\n")
response_mr = await orchestrator.run(query_mr, session_id="demo_mr")
print(response_mr)
```

```python
# Demo 5: Composite Multi-intent query in English
query_en = "I have 1.8 hectares of land in Tumkur. What is today's mandi price for soybean? Also check my PM-KISAN eligibility."
print(f"User Query (en): {query_en}\n")
response_en = await orchestrator.run(query_en, session_id="demo_en")
print(response_en)
```

## 4. Run Evaluation Suite

We can trigger our evaluation harness right from the notebook to display aggregate test metrics.

```python
from evals.run_evals import Evaluator
import json

# Run evaluations on the 100 test cases
evaluator = Evaluator()
await evaluator.run_all()

# Load and print results summary
with open("../evals/eval_results.json", "r", encoding="utf-8") as f:
    results = json.load(f)

print("\n===== EVAL SUMMARY =====")
print(f"Total Test Cases: {results['total_cases']}")
print(f"Passed: {results['passed']}")
print(f"Failed: {results['failed']}")
print(f"Accuracy Percentage: {results['accuracy_pct']}%\n")
```
